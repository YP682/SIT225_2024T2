```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import os

# Check if the input CSV file exists
input_file = 'sensor_data.csv'
if not os.path.exists(input_file):
    print(f"Error: {input_file} does not exist. Make sure the file is in the correct directory.")
else:
    print(f"Found {input_file}, proceeding with data cleaning and plotting.")

    # Load the data
    df = pd.read_csv(input_file)
    print("Original column names in the DataFrame:")
    print(df.columns)
    
    # Clean column names by stripping leading/trailing spaces
    df.columns = df.columns.str.strip()
    print("Cleaned column names in the DataFrame:")
    print(df.columns)

    print("First few rows of the DataFrame:")
    print(df.head())

    # Handle missing values based on actual column names
    df['Temperature'] = pd.to_numeric(df['Temperature'], errors='coerce')
    df['Humidity'] = pd.to_numeric(df['Humidity'], errors='coerce')

    df['Temperature'] = df['Temperature'].fillna(df['Temperature'].mean())
    df['Humidity'] = df['Humidity'].fillna(df['Humidity'].mean())

    # Verify that there are no more NaN values
    print("Number of NaN values in each column after filling:")
    print(df.isna().sum())

    # Save the cleaned data to a new CSV file
    output_file = 'cleaned_sensor_data.csv'
    df.to_csv(output_file, index=False)
    print(f"Cleaned data saved to {output_file}")

    # Plot the data
    plt.figure()
    plt.plot(df['Timestamp'], df['Temperature'], label='Temperature')
    plt.xlabel('Timestamp')
    plt.ylabel('Temperature (°C)')
    plt.title('Temperature Over Time')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    plt.figure()
    plt.plot(df['Timestamp'], df['Humidity'], label='Humidity')
    plt.xlabel('Timestamp')
    plt.ylabel('Humidity (%)')
    plt.title('Humidity Over Time')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    plt.figure()
    plt.plot(df['Timestamp'], df['Temperature'], label='Temperature')
    plt.plot(df['Timestamp'], df['Humidity'], label='Humidity')
    plt.xlabel('Timestamp')
    plt.ylabel('Values')
    plt.title('Temperature and Humidity Over Time')
    plt.legend()
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    print("Data cleaning and plotting complete.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the cleaned data
df = pd.read_csv('cleaned_sensor_data.csv', skip_blank_lines=True)

# Strip column names to remove any potential leading/trailing spaces
df.columns = df.columns.str.strip()

# Check for any invalid data in the Timestamp column
print("Initial few rows of data:")
print(df.head())

# Ensure the Timestamp column is parsed as datetime, skipping errors
df['Timestamp'] = pd.to_datetime(df['Timestamp'], errors='coerce')

# Drop rows with invalid timestamps
df = df.dropna(subset=['Timestamp'])

# Print the first few rows to verify correct parsing
print("Data after parsing Timestamps:")
print(df.head())

# Plot Temperature Over Time
plt.figure(figsize=(10, 5))
plt.plot(df['Timestamp'], df['Temperature'], label='Temperature', color='tab:blue')
plt.xlabel('Timestamp')
plt.ylabel('Temperature (°C)')
plt.title('Temperature Over Time')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('temperature_plot.png')  # Save the temperature plot
plt.show()

# Plot Humidity Over Time
plt.figure(figsize=(10, 5))
plt.plot(df['Timestamp'], df['Humidity'], label='Humidity', color='tab:orange')
plt.xlabel('Timestamp')
plt.ylabel('Humidity (%)')
plt.title('Humidity Over Time')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('humidity_plot.png')  # Save the humidity plot
plt.show()

# Combined Plot for Temperature and Humidity Over Time
plt.figure(figsize=(10, 5))
plt.plot(df['Timestamp'], df['Temperature'], label='Temperature', color='tab:blue')
plt.plot(df['Timestamp'], df['Humidity'], label='Humidity', color='tab:orange')
plt.xlabel('Timestamp')
plt.ylabel('Values')
plt.title('Temperature and Humidity Over Time')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('combined_plot.png')  # Save the combined plot
plt.show()

print("Data plotting complete. Graphs saved as images.")

```
